```python
import pandas as pd 
import seaborn as sns
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure
%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None

#read the data
df = pd.read_csv(r'D:\Alex\python project\movies.csv')
```

```python
df.head(5)
```

```python
#percent of missing data 
for col in df.columns :
    percent_mis = np.mean(df[col].isnull())
    print('{} : {}%'.format(col,percent_mis))
```

```python
df.dtypes
```

```python
#change data types
df.fillna({'gross':0}, inplace=True)
df['gross']=df['gross'].astype('int64')

df.fillna({'votes':0}, inplace=True)
df['votes']=df['votes'].astype('int64')

df.fillna({'budget':0}, inplace=True)
df['budget']=df['budget'].astype('int64')
```

```python
df.sort_values(by='gross', ascending=False )
```

```python
pd.set_option('display.max_rows',None)
df.head()
```

```python
#Drop Duplicates 
df['company'] = df['company'].drop_duplicates()
df['company'].sort_values(ascending=False).head()
```

```python
# sactter plot budget vs score
plt.scatter(x=df['score'],y=df['budget'],color ="black")
plt.title(' budget vs score')
plt.xlabel('score')
plt.ylabel('budget')
plt.show()
```

```python
df.head()
```

```python
#plot budget vs score using seaborn 
sns.regplot(x='score', y='budget', data=df,scatter_kws={"color":"black"}, line_kws={"color":"red"})
plt.show()
```

```python
#see correlation 
numeric_df = df.select_dtypes(include='number')
corr_matrix = numeric_df.corr(method="spearman")
print(corr_matrix)
sns.heatmap(corr_matrix ,annot = True)
plt.title(' correlation matrix heat map', color="purple")
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
corr_matrix.unstack()
corr_sa = corr_matrix.unstack()
sta_corr = corr_sa.sort_values()
sta_corr.head()
```

```python
high_correaltion = sta_corr [(sta_corr) > 0]
high_correaltion.head()
```
